```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Data
features = ['Activity', 'Audio', 'Conversation', 'Location', 'Darkness', 'Charge', 'Lock']
svm_balanced_accuracy = [0.758642, 0.694133, 0.66363, 0.77046, 0.726688, 0.68994, 0.642544]
xgb_balanced_accuracy = [0.853482, 0.774339, 0.806867, 0.843965, 0.841794, 0.807911, 0.823795]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(features, svm_balanced_accuracy, marker='o', linestyle='-', color='blue', label='SVM')
plt.plot(features, xgb_balanced_accuracy, marker='o', linestyle='-', color='orange', label='XGB')

# Add labels and title
plt.xlabel('Feature')
plt.ylabel('Balanced Accuracy')
plt.title('Balanced Accuracy of SVM and XGB Models for Each Feature')
plt.legend()

# Show plot
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Data
features = ['Activity', 'Audio', 'Conversation', 'Location', 'Darkness', 'Charge', 'Lock']
svm_balanced_accuracy = [0.758642, 0.694133, 0.66363, 0.77046, 0.726688, 0.68994, 0.642544]
xgb_balanced_accuracy = [0.853482, 0.774339, 0.806867, 0.843965, 0.841794, 0.807911, 0.823795]

# Plot
bar_width = 0.35
index = np.arange(len(features))

plt.figure(figsize=(10, 6))
plt.bar(index, svm_balanced_accuracy, bar_width, label='SVM', color='blue')
plt.bar(index + bar_width, xgb_balanced_accuracy, bar_width, label='XGB', color='orange')

# Add labels and title
plt.xlabel('Feature')
plt.ylabel('Balanced Accuracy')
plt.title('Balanced Accuracy of SVM and XGB Models for Each Feature')
plt.xticks(index + bar_width / 2, features, rotation=45)
plt.legend()
plt.ylim(0.64, 0.86)
# Show plot
plt.tight_layout()
plt.show()
```
