```python
# https://github.com/lowellh315/DPNewStudents/blob/main/Conversation%20Tutorial.ipynb
def score(response):
    """
    input: string representing the student's response to the phq-9 survey
    output: number representing that score on a scale from 0-3
    """
    if response == 'Not at all':
        return 0
    elif response == 'More than half the days':
        return 2
    elif response == 'Several days':
        return 1
    elif response == 'Nearly every day':
        return 3
    # if the response isn't one of these three, leave it as is
    else:
        return response

# convert each response to a score
PHQ9Numerical = PHQ9.map(score)

# add the scores in each row to make a new column with total scores
PHQ9Numerical['total'] = PHQ9Numerical[PHQ9Numerical.select_dtypes(include=['number']).columns].sum(axis=1)
print(PHQ9Numerical)
```

```python
print(PHQ9Numerical.keys())
```

```python
PHQ9Total = PHQ9Numerical[['uid', 'type', 'total']]
PHQ9Total.rename(columns={'uid': 'userId'}, inplace=True)
PHQ9Total.rename(columns={'total': 'PHQ9'}, inplace=True)

# Displaying the new DataFrame
print(PHQ9Total)
```

```python
pre_df = PHQ9Total[PHQ9Total['type'] == 'pre'][['userId', 'PHQ9']]
pre_df.reset_index(inplace=True)
pre_df = pre_df.drop(columns=['index'])

post_df = PHQ9Total[PHQ9Total['type'] == 'post'][['userId', 'PHQ9']]
post_df.reset_index(inplace=True)
post_df = post_df.drop(columns=['index'])
```

```python
pre_df.to_csv('../dataset/PHQ9/PHQ9Pre.csv', index=False)
post_df.to_csv('../dataset/PHQ9/PHQ9Post.csv', index=False)
```

```python
dataClassificationPre = pre_df.copy()
columns = ['PHQ9']

for column in columns:
    dataClassificationPre.loc[((dataClassificationPre[column] > 0)& (dataClassificationPre[column]<=5)), column] = 0
    dataClassificationPre.loc[((dataClassificationPre[column] > 5)), column] = 1

print(dataClassificationPre)
```

```python
dataClassificationPost = post_df.copy()
columns = ['PHQ9']

for column in columns:
    dataClassificationPost.loc[((dataClassificationPost[column] > 0)& (dataClassificationPost[column]<=5)), column] = 0
    dataClassificationPost.loc[((dataClassificationPost[column] > 5)), column] = 1

print(dataClassificationPost)
```

```python
dataClassificationPre.to_csv('../dataset/PHQ9/PHQ9PreClassified.csv', index=False)
dataClassificationPost.to_csv('../dataset/PHQ9/PHQ9PostClassified.csv', index=False)
```
