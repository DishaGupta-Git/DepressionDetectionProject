```python
import numpy as np
import pandas as pd
import os
from datetime import datetime
```

```python
def Merge(csv_files,directory):
    # Initialize an empty list to store DataFrames
    dataframes = []

    # Loop through each CSV file and read it into a DataFrame, then append to the list
    for file in csv_files:
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path)
        print(file[-6:-4])
        df['userId'] = file[-7:-4]
        # print(df)
        dataframes.append(df)

    # Concatenate all DataFrames in the list into one DataFrame
    merged_df = pd.concat(dataframes, ignore_index=True)
    return merged_df
```

## Merge All Activity

```python
# Path to the directory containing your CSV files
activityDirectory = '../dataset/dataset/sensing/activity'

# List all CSV files in the directory
activityCsv_files = [file for file in os.listdir(activityDirectory) if file.endswith('.csv')]
print(activityCsv_files)
```

```python

# Concatenate all DataFrames in the list into one DataFrame
activityMerged_df = Merge(activityCsv_files,activityDirectory)

# Display or do further operations on the merged DataFrame
print(activityMerged_df)
print(activityMerged_df['userId'].unique())
```

```python
activityMerged_df.to_csv('../dataset/Merged/allActivityRaw.csv', index=False)
```
