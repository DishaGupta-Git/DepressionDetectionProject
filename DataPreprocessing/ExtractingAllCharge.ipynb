```python
ChargeData = pd.DataFrame()
ChargeData = pd.read_csv('../dataset/Merged/allChargeRaw.csv')
print(ChargeData)
```

```python
ChargeData = ChargeData.sort_values(by='userId')
ChargeData.reset_index(drop=True, inplace=True)
ChargeData.rename(columns={'end': 'end_timestamp','start': 'start_timestamp'}, inplace=True)
```

```python
# Convert timestamps to datetime
ChargeData['start_timestamp'] = pd.to_datetime(ChargeData['start_timestamp'], unit='s')
ChargeData['end_timestamp'] = pd.to_datetime(ChargeData['end_timestamp'], unit='s')
print(ChargeData)
```

```python
ChargeData['date'] = ChargeData['start_timestamp'].dt.date
ChargeData['time'] = ChargeData['start_timestamp'].dt.time
ChargeData['date'] = pd.to_datetime(ChargeData['date'])
ChargeData['week'] = ChargeData['date'].dt.isocalendar().week
print(ChargeData)
```

```python
# Calculate conversation length
ChargeData['charge_length'] = ChargeData['end_timestamp'] - ChargeData['start_timestamp']
print(ChargeData['charge_length'])
```

```python
ChargeData['charge_length_seconds'] = ChargeData['charge_length'].dt.total_seconds()
print(ChargeData['charge_length_seconds'])
```

```python
# Convert 'time' column to datetime format
ChargeData['hour'] = ChargeData['start_timestamp'].dt.hour
# print(ActivityData)

bins = [-1, 6, 12, 18, 24]
labels = ['EarlyMorning', 'Morning', 'Afternoon', 'Evening']
ChargeData['timeCategory'] = pd.cut(ChargeData['hour'], bins=bins, labels=labels, right=False)
ChargeData.drop(columns=['hour'], inplace=True)

print(ChargeData)
```

```python
ChargeData = ChargeData.sort_values(by='week')
```

```python
weeks = ChargeData['week'].unique()
# print(weeks)
# print(len(weeks))
# print(weeks[9])
dfs = []

for week in weeks[:-1]:
    twoWeekData = ChargeData[(ChargeData['week'] == week) | (ChargeData['week'] == week+1)]

    # Group by 'timeCategory' and calculate the sum and mean of 'conversation_length_seconds' column
    summary_df = twoWeekData.groupby(['userId','timeCategory'])['charge_length_seconds'].agg(['sum', 'mean']).reset_index()
    summary_df.columns = ['userId','timeCategory', 'Sum', 'Mean']
    # print(summary_df)

    pivot_df = summary_df.pivot_table(index='userId', columns=['timeCategory'], values=['Sum', 'Mean']).fillna(0)
    pivot_df.columns = [f"Charge{category}{calc}" for calc, category in pivot_df.columns]
    # print(pivot_df)

    # Group by 'timeCategory' and find the indices of the shortest and longest conversations
    shortest_indices = twoWeekData.groupby(['userId','timeCategory'], observed=True)['charge_length_seconds'].idxmin()
    longest_indices = twoWeekData.groupby(['userId','timeCategory'], observed=True)['charge_length_seconds'].idxmax()

    # Retrieve the rows corresponding to the shortest and longest conversations
    shortest_conversations = twoWeekData.loc[shortest_indices]
    longest_conversations = twoWeekData.loc[longest_indices]

    ShortestRestructured = shortest_conversations.pivot_table(index='userId', columns=['timeCategory'], values='charge_length_seconds').fillna(0)
    ShortestRestructured.columns = [f"ChargeShortest{category}" for category in ShortestRestructured.columns]
    # print(ShortestRestructured)

    LongestRestructured = longest_conversations.pivot_table(index='userId', columns=['timeCategory'], values='charge_length_seconds').fillna(0)
    LongestRestructured.columns = [f"ChargeLongest{category}" for category in LongestRestructured.columns]
    # print(LongestRestructured)
    longestShortest = pd.merge(ShortestRestructured, LongestRestructured, on='userId')

    merged = pd.merge(pivot_df, longestShortest, on='userId')


    merged['WeekId']= week - weeks[0]
    print(merged)
    dfs.append(merged)

result_df = pd.concat(dfs)
result_df.reset_index(inplace=True)
```
