```python
DarknessData = pd.DataFrame()
DarknessData = pd.read_csv('../dataset/Merged/allDarkRaw.csv')
print(DarknessData)
```

```python
DarknessData = DarknessData.sort_values(by='userId')
DarknessData.reset_index(drop=True, inplace=True)
DarknessData.rename(columns={'end': 'end_timestamp','start': 'start_timestamp'}, inplace=True)
```

```python
# Convert timestamps to datetime
DarknessData['start_timestamp'] = pd.to_datetime(DarknessData['start_timestamp'], unit='s')
DarknessData['end_timestamp'] = pd.to_datetime(DarknessData['end_timestamp'], unit='s')
print(DarknessData)
```

```python
DarknessData['date'] = DarknessData['start_timestamp'].dt.date
DarknessData['time'] = DarknessData['start_timestamp'].dt.time
DarknessData['date'] = pd.to_datetime(DarknessData['date'])
DarknessData['week'] = DarknessData['date'].dt.isocalendar().week
print(DarknessData)
```

```python
# Calculate conversation length
DarknessData['darkness_length'] = DarknessData['end_timestamp'] - DarknessData['start_timestamp']
print(DarknessData['darkness_length'])
```

```python
DarknessData['darkness_length_seconds'] = DarknessData['darkness_length'].dt.total_seconds()
print(DarknessData['darkness_length_seconds'])
```

```python
# Convert 'time' column to datetime format
DarknessData['hour'] = DarknessData['start_timestamp'].dt.hour
# print(ActivityData)

bins = [-1, 6, 12, 18, 24]
labels = ['EarlyMorning', 'Morning', 'Afternoon', 'Evening']
DarknessData['timeCategory'] = pd.cut(DarknessData['hour'], bins=bins, labels=labels, right=False)
DarknessData.drop(columns=['hour'], inplace=True)

print(DarknessData)
```

```python
DarknessData = DarknessData.sort_values(by='week')
```

```python
weeks = DarknessData['week'].unique()
# print(weeks)
# print(len(weeks))
# print(weeks[9])
dfs = []

for week in weeks[:-1]:
    twoWeekData = DarknessData[(DarknessData['week'] == week) | (DarknessData['week'] == week+1)]

    # Group by 'timeCategory' and calculate the sum and mean of 'conversation_length_seconds' column
    summary_df = twoWeekData.groupby(['userId','timeCategory'])['darkness_length_seconds'].agg(['sum', 'mean']).reset_index()
    summary_df.columns = ['userId','timeCategory', 'Sum', 'Mean']
    # print(summary_df)

    pivot_df = summary_df.pivot_table(index='userId', columns=['timeCategory'], values=['Sum', 'Mean']).fillna(0)
    pivot_df.columns = [f"Darkness{category}{calc}" for calc, category in pivot_df.columns]
    # print(pivot_df)

    # Group by 'timeCategory' and find the indices of the shortest and longest conversations
    shortest_indices = twoWeekData.groupby(['userId','timeCategory'], observed=True)['darkness_length_seconds'].idxmin()
    longest_indices = twoWeekData.groupby(['userId','timeCategory'], observed=True)['darkness_length_seconds'].idxmax()

    # Retrieve the rows corresponding to the shortest and longest conversations
    shortest_conversations = twoWeekData.loc[shortest_indices]
    longest_conversations = twoWeekData.loc[longest_indices]

    ShortestRestructured = shortest_conversations.pivot_table(index='userId', columns=['timeCategory'], values='darkness_length_seconds').fillna(0)
    ShortestRestructured.columns = [f"DarknessShortest{category}" for category in ShortestRestructured.columns]
    # print(ShortestRestructured)

    LongestRestructured = longest_conversations.pivot_table(index='userId', columns=['timeCategory'], values='darkness_length_seconds').fillna(0)
    LongestRestructured.columns = [f"DarknessLongest{category}" for category in LongestRestructured.columns]
    # print(LongestRestructured)
    longestShortest = pd.merge(ShortestRestructured, LongestRestructured, on='userId')

    merged = pd.merge(pivot_df, longestShortest, on='userId')


    merged['WeekId']= week - weeks[0]
    print(merged)
    dfs.append(merged)

result_df = pd.concat(dfs)
result_df.reset_index(inplace=True)
```
