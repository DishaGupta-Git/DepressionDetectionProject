```python
LockData = pd.DataFrame()
LockData = pd.read_csv('../dataset/Merged/allLockRaw.csv')
print(LockData)
```

```python
LockData = LockData.sort_values(by='userId')
LockData.reset_index(drop=True, inplace=True)
LockData.rename(columns={'end': 'end_timestamp','start': 'start_timestamp'}, inplace=True)
```

```python
# Convert timestamps to datetime
LockData['start_timestamp'] = pd.to_datetime(LockData['start_timestamp'], unit='s')
LockData['end_timestamp'] = pd.to_datetime(LockData['end_timestamp'], unit='s')
print(LockData)
```

```python
LockData['date'] = LockData['start_timestamp'].dt.date
LockData['time'] = LockData['start_timestamp'].dt.time
LockData['date'] = pd.to_datetime(LockData['date'])
LockData['week'] = LockData['date'].dt.isocalendar().week
print(LockData)
```

```python
# Calculate conversation length
LockData['lock_length'] = LockData['end_timestamp'] - LockData['start_timestamp']
print(LockData['lock_length'])
```

```python
LockData['lock_length_seconds'] = LockData['lock_length'].dt.total_seconds()
print(LockData['lock_length_seconds'])
```

```python
# Convert 'time' column to datetime format
LockData['hour'] = LockData['start_timestamp'].dt.hour
# print(ActivityData)

bins = [-1, 6, 12, 18, 24]
labels = ['EarlyMorning', 'Morning', 'Afternoon', 'Evening']
LockData['timeCategory'] = pd.cut(LockData['hour'], bins=bins, labels=labels, right=False)
LockData.drop(columns=['hour'], inplace=True)

print(LockData)
```

```python
LockData = LockData.sort_values(by='week')
```

```python
weeks = LockData['week'].unique()
# print(weeks)
# print(len(weeks))
# print(weeks[9])
dfs = []

for week in weeks[:-1]:
    twoWeekData = LockData[(LockData['week'] == week) | (LockData['week'] == week+1)]

    # Group by 'timeCategory' and calculate the sum and mean of 'conversation_length_seconds' column
    summary_df = twoWeekData.groupby(['userId','timeCategory'])['lock_length_seconds'].agg(['sum', 'mean']).reset_index()
    summary_df.columns = ['userId','timeCategory', 'Sum', 'Mean']
    # print(summary_df)

    pivot_df = summary_df.pivot_table(index='userId', columns=['timeCategory'], values=['Sum', 'Mean']).fillna(0)
    pivot_df.columns = [f"Lock{category}{calc}" for calc, category in pivot_df.columns]
    # print(pivot_df)

    # Group by 'timeCategory' and find the indices of the shortest and longest conversations
    shortest_indices = twoWeekData.groupby(['userId','timeCategory'], observed=True)['lock_length_seconds'].idxmin()
    longest_indices = twoWeekData.groupby(['userId','timeCategory'], observed=True)['lock_length_seconds'].idxmax()

    # Retrieve the rows corresponding to the shortest and longest conversations
    shortest_conversations = twoWeekData.loc[shortest_indices]
    longest_conversations = twoWeekData.loc[longest_indices]

    ShortestRestructured = shortest_conversations.pivot_table(index='userId', columns=['timeCategory'], values='lock_length_seconds').fillna(0)
    ShortestRestructured.columns = [f"LockShortest{category}" for category in ShortestRestructured.columns]
    # print(ShortestRestructured)

    LongestRestructured = longest_conversations.pivot_table(index='userId', columns=['timeCategory'], values='lock_length_seconds').fillna(0)
    LongestRestructured.columns = [f"LockLongest{category}" for category in LongestRestructured.columns]
    # print(LongestRestructured)
    longestShortest = pd.merge(ShortestRestructured, LongestRestructured, on='userId')

    merged = pd.merge(pivot_df, longestShortest, on='userId')


    merged['WeekId']= week - weeks[0]
    print(merged)
    dfs.append(merged)

result_df = pd.concat(dfs)
result_df.reset_index(inplace=True)
```
