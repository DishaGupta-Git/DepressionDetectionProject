```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#only merge phq9 for baseline assessment - as this has the most data
PHQ9BaseDf = pd.DataFrame()
PHQ9BaseDf = pd.read_csv('Paper33SupData/phq9_baseline.csv')
PHQ9BaseDf.rename(columns={'PHQ9': 'PHQ9Base'}, inplace=True)
print(PHQ9BaseDf)
```

```python
# Start by only merging the 2 week but for weekedn, weekday and all - this has more data points
All2Week = pd.DataFrame()
All2Week = pd.read_csv('Paper33SupData/features_2week_1_all.csv') 
Weekday2Week = pd.DataFrame()
Weekday2Week = pd.read_csv('Paper33SupData/features_2week_2_weekday.csv') 
Weekend2Week = pd.DataFrame()
Weekend2Week = pd.read_csv('Paper33SupData/features_2week_2_weekend.csv') 
```

```python
mergedAll2Week = pd.merge(All2Week, PHQ9BaseDf, on='id')
print(mergedAll2Week)
```

```python
mergedWeekday2Week = pd.merge(Weekday2Week, PHQ9BaseDf, on='id')
print(mergedWeekday2Week)
```

```python
mergedWeekend2Week = pd.merge(Weekend2Week, PHQ9BaseDf, on='id')
print(mergedWeekend2Week)
```

```python
DataCombinedTemp = pd.concat([mergedAll2Week, mergedWeekday2Week], ignore_index=True)
DataCombined = pd.concat([DataCombinedTemp, mergedWeekend2Week], ignore_index=True)
print(DataCombined)
```

```python
savePath = "Paper33SupData/merged/Merged2WeekSet1" 
DataCombined.to_csv(savePath, index=False)
```
