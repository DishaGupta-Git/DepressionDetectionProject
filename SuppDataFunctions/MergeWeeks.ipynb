```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PHQ9BaseDf = pd.DataFrame()
PHQ9BaseDf = pd.read_csv('../Paper33SupData/phq9_baseline.csv')
print(PHQ9BaseDf)
```

```python
# Path to the directory containing your CSV files
directory = '../Paper33SupData'

# List all CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]
print(csv_files)
```

```python
dataframes = []

# Loop through each CSV file and read it into a DataFrame, then append to the list
for file in csv_files:
    if (file[-7:-4] == 'all') and (file[9] == '2'):
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path)
        df.rename(columns={'id': 'userId'}, inplace=True)
        df['WeekId'] = file[-9]
        print(df)
        dataframes.append(df)

# Concatenate all DataFrames in the list into one DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)
```

```python
print(merged_df['userId'].unique())
print(merged_df['WeekId'].unique())
print(merged_df)
```

```python
merged_df = merged_df.drop(columns=['CircadianMovement','NumberofClusters','Entropy','NormalizedEntropy','ContinuousEntropy','TransitionTime' ])
```

```python
merged_df.columns = ['Location' + col if col not in ['userId', 'WeekId'] else col for col in merged_df.columns]
print(merged_df)
```

```python
merged_df.to_csv('../Paper33SupData/merged/AllMerged.csv', index=False)
```
