```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PHQ9Post = pd.DataFrame()
PHQ9Post = pd.read_csv('../dataset/PHQ9/PHQ9PostClassified.csv')
print(PHQ9Post)
```

```python
value_counts = PHQ9Post['PHQ9'].value_counts()

# Plotting the bar chart
plt.bar(value_counts.index, value_counts.values)

# Adding labels and title
plt.xlabel('PHQ9 Score')
plt.ylabel('Frequency')
plt.title('Distribution of Depression Classification')

# Adding x-axis labels
plt.xticks(value_counts.index, ['0', '1'])

# Display the plot
plt.show()
```
