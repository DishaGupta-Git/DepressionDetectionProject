```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Activity Exploration

```python
PHQ9Pre = pd.DataFrame()
PHQ9Pre = pd.read_csv('../dataset/PHQ9/PHQ9Pre.csv')
print(PHQ9Pre)
```

```python
ActivityData = pd.DataFrame()
ActivityData = pd.read_csv('../dataset/BasicFeatures/activityCounts.csv')
print(ActivityData)
```

```python
activityMerged = pd.merge(ActivityData, PHQ9Pre, on='userId')
print(activityMerged)
```

```python
activityMerged = activityMerged.drop(columns=['userId','WeekId'])
```

```python
correlation_matrix = activityMerged.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='PiYG', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')

plt.show()
```
